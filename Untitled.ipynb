```python
import pandas as pd
from matplotlib import pyplot as plt
from scipy.stats import linregress
from sklearn import linear_model
import statsmodels.api as sm



demo_data = pd.read_csv("nc_county_data_clean.csv")
cases_data = pd.read_csv("data/CountyCasesDeathsAndPer10KResidents_Extract_MigratedData.csv")

demo_data = demo_data.rename(columns={"Unnamed: 0": "County"})
demo_data['County'] = demo_data['County'].astype(str).str[:-7]
demo_data = demo_data[['County', 'Population, Census, April 1, 2010', 'Persons under 18 years', 'Persons 65 years and over', 'Female persons', 'White', 'Black or African American', 'American Indian and Alaska Native ', 'Hispanic or Latino', 'Median household income (in 2018 dollars), 2014-2018', 'Persons in poverty', 'High school graduate or higher, 2014-2018', "Bachelor's degree or higher, 2014-2018"]]
df = pd.merge(demo_data, cases_data, how="left", on=["County", "County"])

df.dtypes

for x in df:
    if x == "County":
        continue
    if(df[x].dtype == 'object'):
        if df[x].str.contains('%').any():
            df[x] = df[x].str.replace(',', '')
            df[x] = df[x].str.replace('%', '')
            df[x] = df[x].astype('float')
            df[x] = df[x]/100
            continue
        df[x] = df[x].str.replace('$', '')
        df[x] = df[x].str.replace('.', '')
        df[x] = df[x].str.replace(',', '')
        df[x] = df[x].astype('float')

    
#df['Population, Census, April 1, 2010'] = df['Population, Census, April 1, 2010'].str.replace(',', '').astype('float')
#df['Persons under 18 years'] = df['Persons under 18 years'].str.replace(',', '').astype('float')






#df = df.sort_values('Cases Per 10000 Residents', ascending=True)
df.head(50)
```

```python
# Print out the r-squared value along with the plot.
x_values = df['Cases Per 10000 Residents']
y_values = df['Population, Census, April 1, 2010']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Cases Per 10000 Residents')
plt.ylabel('Population')
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
